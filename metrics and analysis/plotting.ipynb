```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

## Reward Mean Plotting :

```python
file_ppo = r"rewarddata\2025-12-02_16h54m02s_logs_PPO_1.csv"
file_sac = r"rewarddata\2025-12-02_17h01m35s_logs_SAC_1.csv"
```

```python
df_ppo = pd.read_csv(file_ppo)
df_sac = pd.read_csv(file_sac)
```

```python
def plot_comparison():

    df_ppo = pd.read_csv(file_ppo)
    df_sac = pd.read_csv(file_sac)
    plt.figure(figsize=(12, 7))
    plt.style.use('seaborn-v0_8-whitegrid')

    plt.plot(
        df_ppo['Step'], 
        df_ppo['Value'], 
        label='PPO (Proximal Policy Optimization)', 
        color='#1f77b4', 
        linewidth=2, 
        marker='o',     
        markersize=4,
        alpha=0.9
    )

    plt.plot(
        df_sac['Step'], 
        df_sac['Value'], 
        label='SAC (Soft Actor-Critic)', 
        color='#2ca02c', 
        linewidth=2, 
        linestyle='--',  
        marker='x',     
        markersize=5,
        alpha=0.9
    )

    plt.title("Comparative Analysis: Convergence Speed (PPO vs. SAC)", fontsize=16, pad=15)
    plt.xlabel("Training Timesteps (Environment Interactions)", fontsize=12)
    plt.ylabel("Mean Episode Reward", fontsize=12)
    
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    
    plt.legend(fontsize=12, frameon=True, shadow=True)

    output_filename = "ppo_vs_sac_comparison.png"
    plt.savefig(output_filename, dpi=300, bbox_inches='tight')
    plt.show()

if __name__ == "__main__":
    plot_comparison()
```
